# Israeli Construction Equipment Market Analysis

Covers new and used machine registrations from **2018 to April 2026**.

> **Note:** 2026 data is partial (January–April only). Exclude it or annualise when comparing full years.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

from src.loader import load_annual
from src.analysis import (
    total_by_year,
    market_share_by_manufacturer,
    market_share_by_category,
    trend_by_manufacturer,
    top_models,
    import_type_split,
    yoy_growth,
)

sns.set_theme(style='whitegrid', palette='tab10')
FULL_YEARS = list(range(2018, 2026))  # exclude partial 2026

df = load_annual()
print(f'Loaded {len(df):,} rows | years: {sorted(df.year.unique())}')
```

## 1. Market size over time

```python
totals = total_by_year(df)
print(totals)

ax = totals.loc[FULL_YEARS, ['new', 'used']].plot(
    kind='bar', stacked=True, figsize=(10, 5),
    color=['steelblue', 'coral'], edgecolor='white'
)
ax.set_title('Israeli CE Market – Units Registered per Year', fontsize=14)
ax.set_xlabel('')
ax.set_ylabel('Units')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/charts/market_size_by_year.png', dpi=150)
plt.show()
```

## 2. Year-over-year growth (new machines)

```python
growth = yoy_growth(df, condition='new')
print(growth)
```

## 3. Market share by manufacturer (new machines, full years)

```python
mfr_share = market_share_by_manufacturer(df, condition='new', years=FULL_YEARS)
top10 = mfr_share.head(10)
print(top10)

fig, ax = plt.subplots(figsize=(8, 5))
ax.barh(top10['manufacturer_en'][::-1], top10['market_share_%'][::-1], color='steelblue')
ax.set_xlabel('Market Share (%)')
ax.set_title('Top 10 Manufacturers by New Unit Sales (2018–2025)')
for i, (_, row) in enumerate(top10[::-1].iterrows()):
    ax.text(row['market_share_%'] + 0.1, i, f"{row['market_share_%']}%", va='center', fontsize=9)
plt.tight_layout()
plt.savefig('../outputs/charts/manufacturer_share.png', dpi=150)
plt.show()
```

## 4. Equipment category breakdown

```python
cat_share = market_share_by_category(df, condition='new', years=FULL_YEARS)
print(cat_share.head(15))
```

## 5. Manufacturer trend over time (select a category)

```python
CATEGORY = 'EXCAVATORS'   # change to any category_en value

trend = trend_by_manufacturer(df, category_en=CATEGORY, condition='new', top_n=6)
ax = trend.loc[FULL_YEARS].plot(kind='line', marker='o', figsize=(11, 5))
ax.set_title(f'{CATEGORY} – Top Manufacturers (new units)')
ax.set_ylabel('Units')
ax.set_xlabel('Year')
plt.xticks(FULL_YEARS, rotation=0)
plt.tight_layout()
plt.savefig(f'../outputs/charts/trend_{CATEGORY.lower().replace(" ", "_")}.png', dpi=150)
plt.show()
```

## 6. Top models

```python
models = top_models(df, condition='new', years=FULL_YEARS, top_n=15)
print(models)
```

## 7. Commercial vs personal imports

```python
imp = import_type_split(df, years=FULL_YEARS)
print(imp)

ax = imp.plot(kind='bar', figsize=(10, 4), edgecolor='white')
ax.set_title('Commercial vs Personal Import – Units per Year')
ax.set_xlabel('')
ax.set_ylabel('Units')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/charts/import_type_split.png', dpi=150)
plt.show()
```
